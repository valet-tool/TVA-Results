```python
import pandas as pd
```

```python
data_MLP = pd.read_csv("predictions_MLP.csv")
data_LSTM = pd.read_csv("predictions_LSTM.csv")
data_SVR_RBF = pd.read_csv("predictions_SVR_RBF.csv")
```

```python
data_SVR_RBF.head()
```

```python
# if (data_MLP['Tactic__1'] == 1):
#     data_MLP_server1_tactic_1 = data_MLP

data_MLP_server1_tactic_1 = data_MLP[(data_MLP['Tactic__1'] == 1) & (data_MLP['Server__1'] == 1)]
data_MLP_server2_tactic_1 = data_MLP[(data_MLP['Tactic__1'] == 1) & (data_MLP['Server__2'] == 1)]
data_MLP_server3_tactic_1 = data_MLP[(data_MLP['Tactic__1'] == 1) & (data_MLP['Server__3'] == 1)]

data_LSTM_server1_tactic_1 = data_LSTM[(data_LSTM['Tactic__1'] == 1) & (data_LSTM['Server__1'] == 1)]
data_LSTM_server2_tactic_1 = data_LSTM[(data_LSTM['Tactic__1'] == 1) & (data_LSTM['Server__2'] == 1)]
data_LSTM_server3_tactic_1 = data_LSTM[(data_LSTM['Tactic__1'] == 1) & (data_LSTM['Server__3'] == 1)]

data_SVR_RBF_server1_tactic_1 = data_SVR_RBF[(data_SVR_RBF['Tactic__1'] == 1) & (data_SVR_RBF['Server__1'] == 1)]
data_SVR_RBF_server2_tactic_1 = data_SVR_RBF[(data_SVR_RBF['Tactic__1'] == 1) & (data_SVR_RBF['Server__2'] == 1)]
data_SVR_RBF_server3_tactic_1 = data_SVR_RBF[(data_SVR_RBF['Tactic__1'] == 1) & (data_SVR_RBF['Server__3'] == 1)]
```

```python
data_MLP_server1_tactic_1.to_csv('prediction_MLP_server1_tactic_1.csv')
data_MLP_server2_tactic_1.to_csv('prediction_MLP_server2_tactic_1.csv')
data_MLP_server3_tactic_1.to_csv('prediction_MLP_server3_tactic_1.csv')

data_LSTM_server1_tactic_1.to_csv('prediction_LSTM_server1_tactic_1.csv')
data_LSTM_server2_tactic_1.to_csv('prediction_LSTM_server2_tactic_1.csv')
data_LSTM_server3_tactic_1.to_csv('prediction_LSTM_server3_tactic_1.csv')

data_SVR_RBF_server1_tactic_1.to_csv('prediction_SVR_RBF_server1_tactic_1.csv')
data_SVR_RBF_server2_tactic_1.to_csv('prediction_SVR_RBF_server2_tactic_1.csv')
data_SVR_RBF_server3_tactic_1.to_csv('prediction_SVR_RBF_server3_tactic_1.csv')

```
